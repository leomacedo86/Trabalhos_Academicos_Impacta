## IMPORTS

```python
import requests
import pandas as pd
import seaborn as sns
from matplotlib            import pyplot as plt
```

### MARCA  1 - MAYBELLINE

```python
response_maybelline = requests.get('http://makeup-api.herokuapp.com/api/v1/products.json?brand=maybelline')
```

```python
response_maybelline.status_code
```

```python
df_maybelline = pd.DataFrame(response_maybelline.json())
df_maybelline.head()
```

```python
df_maybelline.to_csv('Maybelline.csv')
```

## UNINDO OS DATAFRAMES

```python
df_marcas = pd.concat([df_maybelline, df_annabelle, df_revlon, df_clinique,df_benefit])
```

```python
df_marcas['brand_name'] = df_marcas ['brand']
```

```python
df_marcas.head()
```

```python
aux1 = df_marcas['brand'].value_counts().reset_index()
```

```python
df_marcas_sum = pd.DataFrame (aux1)
```

```python
df_marcas_sum.head()
```

```python
df_marcas_sum.columns = ['nome_marca','total_produtos']
```

```python
df_marcas_sum.head()
```

# PLOTANDO O GR√ÅFICO

```python
sns.barplot( x='nome_marca', y='total_produtos', data=df_marcas_sum)
```
